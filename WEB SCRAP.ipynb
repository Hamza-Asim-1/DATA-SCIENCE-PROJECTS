```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#defining the url
url='https://www.daftlogic.com/information-appliance-power-consumption.htm'
```

```python
web=requests.get(url)
s=BeautifulSoup(web.content,'html.parser')
```

```python
#finding the content using table id and storing the data into a dataframe
content = s.findAll(attrs = {'id': 'tblApp'})
df = pd.read_html(str(content))[0]
df
```

```python
#converting dataframe to csv file
df.to_csv('appliances.csv')
```
