# Life Expectancy (WHO)<a href="#Life-Expectancy-(WHO)" class="anchor-link">¶</a>

The dataset, compiled from the Global Health Observatory (GHO) data
repository under the World Health Organization (WHO) and the United
Nations, encompasses health and economic factors for 193 countries from
2000 to 2015. The study focuses on critical variables within the
categories of immunization, mortality, economy, and social factors.
