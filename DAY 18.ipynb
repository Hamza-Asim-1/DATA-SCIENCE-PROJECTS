```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://demoazulmexicanosummerville.onlineordersnow.com/MenuDisplay"
driver = webdriver.Edge()
driver.get(url)
# waiting for the view menu buttin to be presented
view_menu_button = WebDriverWait(driver,30).until(
 EC.presence_of_element_located((By.XPATH, "//div[@class='v-btn__content' and contains(text(), 'View Menu')]")) # El
)
view_menu_button.click()
# waiting for the open menu to be presented
open_menu_button = WebDriverWait(driver,30).until(
 EC.presence_of_element_located((By.XPATH, "//div[@class='v-btn__content' and contains(text(), 'Open Menu')]")) # El
)
open_menu_button.click()
# find all the size and price boxes
size_price_boxes = WebDriverWait(driver,30).until(
 EC.presence_of_all_elements_located(
 (By.XPATH, "//span[@class='sizeStyle mr-1 floatRight' and contains(text(), '(Click for Size/Price)')]")
 )
)
# for box in size_price_boxes: 
# box.click()
# Get the page source
html = driver.page_source
# Close the driver
driver.close()

```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'lxml')
item1 = soup.find_all('div', class_='flex pb-1 md12 xs12')

item1_menu = []

for i in item1:
    item_text = i.select_one('span.itemLabel').text
    item_price = i.select('span')[-1].text
    item1_menu.append([item_text, item_price])

# Converting list to a DataFrame
item1_df = pd.DataFrame(item1_menu, columns=['item', 'price'])

# Save DataFrame to a CSV file
item1_df.to_csv('item1_menu.csv', index=False)

print(item1_df)
```
